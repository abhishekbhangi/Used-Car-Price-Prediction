```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("D:\MLP\Multi Linear Regression\cardekho_imputated.csv")
df
```

```python
df.describe()
```

```python
X = df.iloc[:,[4,5,9,10,11,12]]
y= df.iloc[:,13]
```

```python
X.head()
```

```python
y.head(20)
```

```python
### Train Test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=35)
print(X_train,X_test,y_train,y_test)
```

```python
### Implement Random Forest classifier
from sklearn.linear_model import LinearRegression
regression=LinearRegression()#model building
regression.fit(X_train,y_train)#model training
```

```python
###prediction
y_pred = regression.predict(X_test)
y_pred
```

```python
### check accuracy
from sklearn.metrics import r2_score,mean_squared_error
score= r2_score(y_test,y_pred),mean_squared_error(y_test,y_pred)
score
```
